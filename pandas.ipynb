O Pandas é uma biblioteca de código aberto em Python usada para manipulação e análise de dados em formato de tabelas, chamadas de DataFrames. Ela oferece ferramentas poderosas para limpeza, transformação e análise de dados.


```python
import pandas as pd

list = [20,40,60,80,100]
serie_pandas = pd.Series(list)

print(serie_pandas)
```

```python
notes = {"Mafe": 10, "Guilherme": 7.8, "Isa": 9.5}
serie_notes = pd.Series(notes)

print(serie_notes['Mafe'])

print("Média:", serie_notes.mean())
print("Mediana:", serie_notes.median())
print("Descrição:", serie_notes.describe())
```

Pandas Dataframes

```python
animes = {"Autores": ["Katsuhiro Otomo", "Shinichiro Watanabe", "Yoshihiro Togashi"],
          "Animes":  ["Akira", "Cowboy Bebop", "HunterxHunter"]}

df = pd.DataFrame(animes)
df

```

```python
df["Autores"]
```

```python
df["Autores"][2]
```

```python
mask = (df["Autores"] == "Yoshihiro Togashi")
print(mask)
```

```python
df[mask]
```

```python
df = df.append({"Autores": "Togashi", "Animes": "Yu Yu Hakusho"}, ignore_index = True)
```

```python
df
```

```python
mask_correct = df["Animes"] == "HunterxHunter"
df.loc[mask_correct, "Animes"] = "Hunter x Hunter"
df
```
